# Planteamiento del problema

En este trabajo buscaremos encontrar cuál es la relación entre el aumento en el uso de energías renovables y la posibilidad de disminuir los niveles de CO<sub>2</sub>.

Parte del problema radica en el rápido agotamiento de las fuentes de energía no renovables: si los países que consumen más energía no modifican sus políticas energéticas la demanda mundial de energía crecerá hasta en 50% para 2030, según estimaciones de la AIE (Agencia Internacional de la Energía). Esto resulta preocupante dado que las reservas de petroleo y gas natural, ambos combustibles no renovables, empiezan a agotarse; según estimaciones el petróleo y el gas natural supondrán un 60% de la demanda de energía en 2030, mientras que las energías renovables representarán tan sólo un 10%[1]. Un artículo reciente muestra que las reservas de petróleo no son suficientes para abastecer el consumo estimado de energía de los próximos 30 años, y las de carbón son apenas suficientes[2]. Según este mismo artículo, si consideramos solamente la generación de energía eléctrica ninguno de estos recursos será suficiente, dado que grán parte de su valor energético proviene de la combustión y este no es eficientemente convertido en electricidad. Ante este escenario, es vital utilizar fuentes alternativas que puedan aprovecharse una vez que estas fuentes hayan sido agotadas.

![Reservas](https://ars.els-cdn.com/content/image/1-s2.0-S266711312200002X-gr2.jpg "Energía disponible")

* **Figura 1.** Cantidad de energía razonablemente asegurada de fuentes renovables y finitas para los próximos 30 años. Cada circulo es proporcional a las reservas energéticas disponibles por tipo de recurso. El circulo de la izquierda representa el consumo energético total estimado para 30 años. Fuente [2].

Las energías renovables son una de las principales alternativas para suplir las necesidades energéticas, pues tal como indica su nombre hace uso de recursos que pueden ser recuperados o restablecidos en un tiempo relativamente corto, principalmente en contraste con los millones de años que tomó la formación de los combustibles fósiles[3]. En esta categoría existen varios tipos de energía dado su orígen, siendo los principales la energía eólica, solar, hidráulica y geotérmica. Para cada uno de estos tipos de energía renovable existes distintos métodos para su aprovechamiento cuya eficiencia varía, además de que algunos tipos son accesibles solamente en lugares específicos o son más eficientes bajo ciertas condiciones. Estas limitaciones, aunadas a los interéses económicos que existen en el uso de combustibles fósiles, explican en parte porqué no ha sido posible sustituir de forma tajante nuestras fuentes energéticas actuales por las renovables [4]. A pesar de esto, desde 2020 el ritmo de crecimiento en el uso de energía renovables se ha acelerado hasta el 12%, gracias en parte al apoyo fiscal de los gobiernos y el aumento de la financiación sostenible. Actualmente las energías renovables representan ahora más del 80% de la inversión total en el sector eléctrico, y el gasto en energía solar fotovoltáica, baterías y vehículos eléctricos está creciendo a un ritmo consistente con el objetivo de alcanzar emisiones cero netas para 2050.[5]


A la par de esta situación se encuentra el problema del cambio climático. El cambio climático es uno de los 23 problemas más importantes para la Organización de las Naciones Unidas, nombrandolo el problema definitorio de nuestros tiempos, además de indicar que estamos en el momento definitorio para abordarlo [6]. En grán medida debido a las emisiones que involucra el consumo de combustibles fósiles, estamos experimentando cambios extremos en las condiciones climáticas de nuestro planeta, lo que provoca fenómenos como sequías, inundaciones, afectaciones en la producción de alimentos y disminución en las poblaciones de algunas especies. [7] Esto incrementa el interés en dejar de utilizar combustibles fósiles, a la vez de que debe considerarse al proponer alternativas energéticas. Actualmente los esfuerzos por combatir el cambio climático son globales, quedando como referente actual el Acuerdo de París [8]. El uso de energías renovables ayudaría de forma importante a desacelerar el cambio climático, considerando que en el proceso de generación no se producen emisiones de gases de efecto invernadero [9].

![Energia renovable y cambio climático](https://www.irena.org/-/media/Images/IRENA/Agency/Topics/Climate-Change/REcrucial_for_climate_change_infographic.jpg "Energía renovable")
* **Figura 2** EL uso de energía renovable es crucial para poder cumplir con las metas del Acuerdo de Paris. Fuente [10].

Existen otras posibles soluciones para el problema del abasto energético y el del cambio climático. Una de las alternativas para el abasto de energía es aumentar el uso de energía nuclear, dado que puede suplir de grandes cantidades de energía y no emite CO<sub>2</sub>[11]. Actualmente las desventajas de esta alternativa son principalmente los costos de operación y la percepción negativa de la población sobre esta energía, dado el antecedente histórico de las bombas nucleares. La fisión nuclear, en contraste con el proceso de fusión nuclear tradicionalmente utilizado en las plantas nucleares, promete ser una fuente de energía importante; lamentablemente se estima que esta tecnología sea utilizable en 20 años más, por lo que no se considera una solución en el urgente combate al cambio climático[12].

Para combatir el cambio climático algunas propuestas son disminuir el consumo de productos ganaderos, la reforestación, protección de grandes bosques como el Amazonas, disminuir la sobrepesca y la contaminación de los oceanos, disminuir el uso de plásticos y cambiar los hábitos actuales de consumo. Todas estas acciones se asocian a disminuir la producción de CO<sub>2</sub> o a reducir la cantidad actualmente presente[13]. 


Dada esta información, es evidente que abordar este problema resulta de vital importancia actualmente, y cualquier esfuerzo para comprenderlo y consecuentemente proponer posibles soluciones resulta de gran valor.

```python
# funciones
def grafica_barras(data, x_axis, y_axis, title='title', x_label='x', y_label='x'):
  sns.set(font_scale = 1.5)
  sns.set_style('white')
  graphic = sns.barplot(
      x=x_axis,
      y=y_axis,
      data=data, 
      palette = 'YlGnBu'
    ).set_title(title, fontsize = 20, weight = 'bold')
  plt.xlabel(x_label, weight = 'bold')
  plt.ylabel(y_label, weight = 'bold')
  graphic.figure.set_size_inches(28,12)
  sns.despine()

def grafica_circular(data, x_axis, y_axis, title='title'):
  labels = data[x_axis]
  sizes = data[y_axis]
  porcent = 100.*sizes/sizes.sum()
  # Colores
  colors = ['#b6d7a8','#ffe599', '#b6b4af', '#e06666', '#d5a6bd','#f6b26b']

  fig1, ax1 = plt.subplots(1, figsize=(15, 12), facecolor='#85C1E9')
  patches, texts = ax1.pie(sizes, colors = colors, startangle=90, wedgeprops={'linewidth':0})

  labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(labels, porcent)]

  sort_legend = True
  if sort_legend:
      patches, labels, dummy =  zip(*sorted(zip(patches, labels, sizes),
                                          key=lambda x: x[2],
                                          reverse=True))
  plt.legend(patches, labels, bbox_to_anchor=(1,0.5), loc="center right", bbox_transform=plt.gcf().transFigure,
           fontsize=13)
  plt.title(title,fontweight='bold', fontsize = 24, color = "white")

  # La misma relación de aspecto garantiza que el pastel se dibuje como un círculo
  ax1.axis('equal')  
  plt.tight_layout()
  plt.show()
```

#### Pregunta 3. ¿Cuál es el continente que produce más energía?

Para la resolución de esta pregunta, lo que haremos será utilizar un dataframe contiene datos sobre 34936 fábricas de diferentes países en el mundo, de donde podemos obtener la capacidad de producción energética de cada fábrica en megavatios, despues se agrupan por pais y finalmente por continente.

```python
# Se filtran lo registros de Estados Unidos
datos_estados_unidos = df_energia_filtrado[df_energia_filtrado['country'] == 'United States']

# Se obtiene el porcentaje de NaN de los datos
datos_estados_unidos['population'].isna().sum() / len(datos_estados_unidos) * 100
```

Se observa que el 50% de los datos son NaN, lo que explica que el país no figure entre los 10 países más poblados obtenida anteriormente. Para solucionar esta situación, los valores NaN se llenaran con el valor de la media del 50% de los datos que no son NaN

```python
# Se obtienen los datos donde 'population' no es NaN
datos_estados_unidos_sin_nan = datos_estados_unidos[datos_estados_unidos['population'].notna()]
datos_estados_unidos_sin_nan
```

```python
# Se obtiene la media de la población 
media_poblacion_estados_unidos = datos_estados_unidos_sin_nan['population'].mean()
```

```python
# Se crea un filtro para obtener los renglones donde el país sea Estados Unidos y la el valor de la población sea NaN
filtro = (df_energia_filtrado['country'] == 'United States') & (df_energia_filtrado['population'].isna())
```

```python
# se realiza una copia del DataFrame para no alterar el original
df_energia_corregido = df_energia_filtrado.copy()

# Se aplica la función mask para remplazar los NaN con el valor de la media
df_energia_corregido['population'] = df_energia_corregido['population'].mask(filtro, df_energia_corregido['population'].fillna(media_poblacion_estados_unidos))

df_energia_corregido
```

```python
# Se vuelve a rellenan los NaN de la columna de interés mediante una interpolación lineal de los datos
df_energia_corregido['population'] = df_energia_corregido['population'].interpolate(method='linear', limit_direction='forward')
df_energia_corregido.isna().sum(axis=0)
```

```python
# Se obtiene la media de la población por país
poblacion_por_pais = df_energia_corregido.groupby('country')['population'].mean()

# Se obtienen nuevamente los 10 países más poblados
poblacion_top_10_paises = poblacion_por_pais.sort_values(ascending = False).head(10).reset_index()
poblacion_top_10_paises
```

Se observa que Estados Unidos ya aparece en la lista, no obstante, según la fuente anterior debería estar en el 3er lugar, por lo que se concluye que la solución anterior no fue suficiente para corregir los datos, debido a la gran cantidad de NaN presentes, 50%.

```python
# Country
#  CO2_emission - The amount of C02 emitted, measured (MMtonnes CO2)
#  Population - Population of specific Country, measured (Mperson)
#  CO2_emission / Population
#  where: Energy_type = "all_energy_types"
#lectura del dataset 2
#df_energia2= df_energiaP9 #copia del dataset general 
df_energia2.rename(columns={'energy_type' : 'e_type', 'co2_emission' : 'co2', 'population' : 'c_pop'
                   }, inplace=True)
  # al filtrar por "all_energy_types" se obtienenen los totales CO2_emission & Population de cada pais 
temp_co2 = df_energia2[df_energia2['e_type']=="all_energy_types"]
  # se eliminan los NaN, es decir los paises sin datos de emisiones de CO2
temp_co2 = temp_co2.dropna(axis=0, subset=['co2'], how = 'any')
  # se quitan columnas extras
temp_co2 = temp_co2.drop(columns=['e_type', 'energy_consumption','gdp','energy_intensity_per_capita','energy_production','energy_intensity_by_gdp'])
# calculo de emision de Co2 per capita / per pais/ per year
temp_co2["co2_capita"] = temp_co2["co2"] / temp_co2["c_pop"] * 1000
# agrupamos por pais y sacamos la media de todos los anios 
ds_co2 = temp_co2.groupby('country')['co2_capita'].agg([pd.Series.mean])
ds_co2.rename(columns={'mean' : 'co2_per_capita'}, inplace=True)
ds_co2 = ds_co2.dropna(axis=0, subset=['co2_per_capita'], how = 'any')
#ordenamos y seleccionamos el top10
ds_co2= ds_co2.sort_values("co2_per_capita", axis = 0, ascending = False).head(10)
ds_co2
```

```python
def plot_co2_econ(dfl, country):
    #obtener registros de emision de CO2 - USA
    ds_co2_1= dfl.loc[ (dfl['Country']==country) & (dfl['e_type']=="all_energy_types") ]
    #ds_co2_1 = ds_co2_1.drop(columns=opciones)
    ds_co2_1 = ds_co2_1.drop(columns=['Country','e_type', 'e_con', 'c_pop','GDP','Energy_intensity_per_capita','Energy_production','Energy_intensity_by_GDP'])
    ds_co2_1.set_index("Year", inplace = True)
    # Ajuste para escalar la grafica vs consumo e-renovable
    ds_co2_1["co2"] = ds_co2_1["co2"] / 100
    #print(ds_co2_1.head())
    #obtener registros de produccion de E-Limpias - USA
    ds_e_con_1= dfl.loc[ (dfl['Country']==country) & (dfl['e_type']=="all_energy_types") ]
    #ds_e_con_1 = ds_e_con_1.drop(columns=opciones)
    ds_e_con_1 = ds_e_con_1.drop(columns=['Country','e_type', 'co2', 'c_pop','GDP','Energy_intensity_per_capita','Energy_production','Energy_intensity_by_GDP'])
    ds_e_con_1.set_index("Year", inplace = True)
    #print(ds_e_con_1.head())
    #plot multiple lines with legend
    plt.plot(ds_e_con_1.index, ds_e_con_1['e_con'], 'g', label='Consumo E-Limpias')
    plt.plot(ds_e_con_1.index, ds_co2_1['co2'], 'r', label='Emision CO2')
    plt.title('Comparacion '+country+': Emision CO2 vs Consumo E-Limpias en el tiempo')
    plt.legend()
    plt.show()
```

```python
plot_co2_econ(df_energia3, 'United States')
```

```python
plot_co2_econ(df_energia3, 'China')
```

```python
plot_co2_econ(df_energia3, 'Canada')
```
